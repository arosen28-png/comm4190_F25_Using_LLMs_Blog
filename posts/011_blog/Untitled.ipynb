The Experiment

We hear a lot about “AI competition” — but what does it *actually* look like when three large language models face the same question?

For this experiment, I posed identical prompts to **ChatGPT**, **Gemini**, and **Perplexity**.  
Each model had the same goal: *explain the ethics of AI creativity in under 100 words.*

Here’s what happened.

---

### **Prompt 1: "Can machines create ethically?"**

- **Gemini** wrote like a philosopher — abstract, balanced, and careful not to offend.  
- **ChatGPT** wrote like a writer — structured, reflective, and humanly cautious.  
- **Perplexity** pulled facts — citing researchers and summarizing public debates.

Each was right in its own way. Together, they revealed how “intelligence” reflects *perspective*.

---

### **Prompt 2: "Describe a future where AIs collaborate with artists."**

Gemini leaned poetic.  
ChatGPT built a narrative.  
Perplexity turned into a curator — listing examples and links.

It made me realize: **the way a model answers tells you what it values.**

---

### **Takeaway**

Testing LLMs isn’t just about accuracy — it’s about *attitude.*  
Each model speaks from a slightly different worldview, like three mirrors reflecting the same question at different angles.

If we learn to listen to all three, we don’t just get better answers — we get better questions.
