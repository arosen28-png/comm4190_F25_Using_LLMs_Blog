In A Brief History of Intelligence, the book talks about how minds evolved through different stages — from reacting, to learning, to eventually imagining.
While reading, I started noticing how prompts seem to “nudge” AI into something similar. Some prompts make the model respond quickly and simply. Others make it think more deeply or creatively.

It made me wonder if prompts act like small switches that move the AI between these different levels.

A simple prompt gets a simple answer.
A more open prompt gets something more layered.

Leveling Up the Thinking

Some prompts stay basic:

“Define this.”

“Summarize that.”

Others ask for more thought:

“Explain why this matters.”

“Compare these ideas.”

And then some prompts push toward imagination:

“Imagine this…”

“What if…?”

The book describes how humans move between different modes of thinking all the time. Prompts sort of do that for AI — they push it up or down these levels.

<h1 style="color: #96174e"> Reflection </h1>

Thinking about the book made me realize prompting isn’t just about wording things the right way.
It’s about choosing what kind of thinking you want to spark.

If I ask small questions, I get small answers.
If I ask bigger or more flexible questions, the AI tends to follow along.

It’s funny — prompting AI is teaching me how I prompt myself too.
