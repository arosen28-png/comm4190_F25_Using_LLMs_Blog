<h1 style="color: #96174e">Iterating Prompts to Improve Blog Summaries</h1>

<p>In this post, I explored how changing and refining prompts affects the quality of summaries produced by an LLM. Summarization is a common use case for blogging, especially when working with long readings, lecture notes, or articles.</p>

<p>Rather than using a single prompt, I tested multiple versions to see how small wording changes influenced the output. The goal was to produce a summary that was accurate, concise, and appropriate for a blog audience.</p>

<h4 style="color:#0707ff">The Original Text</h4>

<p style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
The article discusses how repeated exposure to media messages can shape beliefs over time. It emphasizes that effects are gradual and depend on context, frequency, and audience interpretation rather than immediate persuasion.
</p>

<h4 style="color:#0707ff">Prompt 1</h4>

<div style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
<blockquote>
Summarize this paragraph.
</blockquote>
</div>

<h4 style="color:#0707ff">Response 1</h4>

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 8px;">
<blockquote>
The article explains that media messages influence beliefs gradually over time, depending on exposure and audience interpretation.
</blockquote>
</div>

<p>This response was accurate but very minimal. While it captured the core idea, it lacked detail and felt too generic for a blog post.</p>

<h4 style="color:#0707ff">Prompt 2</h4>

<div style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
<blockquote>
Summarize this paragraph for a blog audience, focusing on the main idea and key factors involved.
</blockquote>
</div>

<h4 style="color:#0707ff">Response 2</h4>

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 8px;">
<blockquote>
The article argues that media messages shape beliefs slowly rather than instantly, with effects depending on repeated exposure, context, and how audiences interpret what they see.
</blockquote>
</div>

<p>This version added more specificity and was better suited for a blog. It highlighted the gradual nature of media influence and included the key factors mentioned in the original text.</p>

<h1 style="color: #96174e">Reflection</h1>

<p>This experiment showed that prompt iteration plays a major role in how useful an LLMâ€™s output is. A vague prompt produced a technically correct but shallow summary, while a slightly more specific prompt led to a clearer and more informative result.</p>

<p>For blogging, this suggests that spending time refining prompts can significantly improve output quality. Instead of accepting the first response, adjusting the prompt helps guide the model toward the intended audience and purpose.</p>

<p>Overall, this use case reinforced that effective LLM use is less about asking once and more about refining questions through iteration.</p>
