<h1 style="color: #96174e">Using LLMs to Control Tone in Blog Writing</h1>

<p>In this post, I examined how an LLM can be used to rewrite the same content in different tones. Tone control is especially important for blogging, since the same idea can sound academic, casual, or reflective depending on how it is written.</p>

<p>Rather than asking the model to create new ideas, I used an existing paragraph and tested how different prompts changed the tone while keeping the meaning consistent.</p>

<h4 style="color:#0707ff">Original Paragraph</h4>

<p style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
Large language models are becoming increasingly common tools for writing and studying. However, their usefulness depends on how intentionally they are used, especially when clarity and audience matter.
</p>

<h4 style="color:#0707ff">Prompt 1: Formal Academic Tone</h4>

<div style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
<blockquote>
Rewrite this paragraph in a formal academic tone.
</blockquote>
</div>

<h4 style="color:#0707ff">Response 1</h4>

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 8px;">
<blockquote>
Large language models have become increasingly prevalent tools in academic and professional writing contexts. Their effectiveness, however, is contingent upon intentional and audience-aware use, particularly when clarity is a priority.
</blockquote>
</div>

<p>This version sounded polished and appropriate for a paper, but it felt distant and less personal.</p>

<h4 style="color:#0707ff">Prompt 2: Casual Blog Tone</h4>

<div style="background-color: #f8f8f8; padding: 10px; border-radius: 8px;">
<blockquote>
Rewrite this paragraph in a casual, conversational blog tone.
</blockquote>
</div>

<h4 style="color:#0707ff">Response 2</h4>

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 8px;">
<blockquote>
LLMs are becoming a normal part of how people write and study, but how useful they are really depends on how thoughtfully they’re used — especially when you’re trying to communicate clearly with a specific audience in mind.
</blockquote>
</div>

<p>This version felt more approachable and better suited for a blog, even though the core idea stayed the same.</p>

<h1 style="color: #96174e">Reflection</h1>

<p>This experiment showed that LLMs are effective tools for adjusting tone without changing meaning. Small prompt changes can dramatically alter how text feels to a reader.</p>

<p>For blogging, this makes LLMs especially useful during revision. Instead of rewriting entire sections manually, I can quickly test different tones and then choose the one that best fits the audience and purpose.</p>

<p>Overall, this use case reinforced that LLMs are not just writing tools, but also powerful revision and style-control tools.</p>
